### Understanding the hyperplane equation

$$\vec{w}=(w_{0}, w_{1})$$
$$\vec{x}=(x, y)$$
$$\vec{w}*\vec{x}+b=0$$


$$ w_{0}*x+w_{1}*y+b=0 $$
$$ y = \frac{-w_{0}}{w_{1}}*x - \frac{b}{w_{1}} $$

$$ a = \frac{-w_{0}}{w_{1}}; c = - \frac{b}{w_{1}}$$

$$ y=a*x-c $$
